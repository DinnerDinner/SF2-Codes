```python
columns = df.columns
data_df = []

# we have 11 runs so i removed 4 from the total len so we only have 10 runs as asked
for i in range(0, len(columns)-4, 4): #time
    data_df.append(columns[i])

for i in range(1,len(columns)-4,4): #position
    data_df.append(columns[i])
    
data_df = df[data_df]
# data_df = data_df.replace(np.nan, 0)
```

6. The data in `data_df` contains noisy data.  In this step you are to clean the noisy data (or extract the clean data). Your cleaned data **for each run**
- must not include the data that is after once the filter hits the motion sensor
- must include the curved data that transitions into a drop in position.  You should only keep about 0.5 to 1 second of the roughly constant position values before the drop of position takes place.  
Note that you should clean your data for each of the 10 runs and you should automate this process in your code with logical conditions, loops, etc.

7. Store your clean data into a new DataFrame called `clean_df`. 
8. Note that now that the values in the **Time** columns may have shifted due to the clean-up.  Update your `clean_df` to rescale all the **Time** columns to start the time from 0.0 in increments of 0.02. 
9. Write this to a new .csv file called `"clean_data_single_filter.csv"`. 

```python
data_df
clean_df = data_df.copy()
columns = data_df.columns

for i in range(10, 20):
    position_col = columns[i]
    
    # cut at end code, first occasion of the smallest value
    zero_index = data_df[position_col].idxmin()
    if pd.notna(zero_index):
        clean_df[position_col] = clean_df[position_col].iloc[:zero_index+1]
    

    # cut at start code, pretty much first 7 descending values indicate the drop
    values = data_df[position_col].values
    for j in range(len(values) - 6):
        window = values[j:j+7]
        if all(window[k] > window[k+1] for k in range(len(window)-1)):
            clean_df[position_col] = clean_df[position_col].iloc[j+7:]
            break

    # #smoothens out my data
    # values = clean_df[position_col].values
    # for j in range(len(values) - 1):
    #     if values[j] <= values[j+1]:
    #         mean_value = (values[j] + values[j+2]) / 2
    #         clean_df[position_col].iloc[j+1] = mean_value
        
    #adjusting time values, pretty much aligns position start with time start
    first_valid_position_index = clean_df[position_col].first_valid_index()
    if first_valid_position_index is not None:
        time_col = columns[i - 10] 
        first_time_value = clean_df[time_col].iloc[first_valid_position_index]
        clean_df[time_col] = clean_df[time_col] - first_time_value 

clean_df.to_csv("clean_data_single_filter.csv", index=False)
```

10. Plot the data using `matplotlib` from all 10 runs as scatter plots on the same figure using your cleaned data.  
- Each plot should have a label **Run #$i$** for $1\leq i\leq10$
- The label on the x-axis should read **Time (s)**
- The label on the y-axis should read **Position (m)**
- Each plot should be in a different color and there should be a legend showing which colour is which run. 
- scale the size of the points so that they are not too big.  
Note that you should automate the process of plotting the 10 graphs.

11. Take the last 0.5 to 1 second of the x-values in your 10 runs, and draw a line of best fit on all your 10 graphs. 

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(24, 16))


for i in range(10, 20):
    position_col = columns[i] 
    time_col = columns[i - 10] 
    

    scatter = plt.scatter(clean_df[time_col], clean_df[position_col], label=f'Run #{i-9}', s=20)


    first_valid_idx = clean_df[position_col].first_valid_index()
    min_idx = clean_df[position_col].idxmin()

    first_valid_pos = clean_df[position_col].loc[first_valid_idx]
    min_pos = clean_df[position_col].loc[min_idx]

    first_valid_time = clean_df[time_col].loc[first_valid_idx]
    min_time = clean_df[time_col].loc[min_idx]

    m = (min_pos - first_valid_pos) / (min_time - first_valid_time)
    b = first_valid_pos - m * first_valid_time
    time_fit = [first_valid_time, min_time]
    pos_fit = [m * first_valid_time + b, m * min_time + b]
    plt.plot(time_fit, pos_fit, linestyle='-', color=scatter.get_facecolor()[0], linewidth=2)


plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.title('Position vs Time for All Runs')
plt.legend()
plt.show()
```
